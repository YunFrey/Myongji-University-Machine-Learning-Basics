# 3) 타이타닉 데이터에서 탑승 항구 변수를 인코딩하기

### 3-1) 타이타닉 데이터 불러오기

```python
df_titanic = pd.read_csv('titanic.csv')
#매핑
df_titanic["Sex"] = df_titanic.Sex.map({"male":1,"female":0})
df_titanic["Embarked"] = df_titanic.Embarked.map({"S":0,"C":1,"Q":2})
#단, 인코딩할 Embarked 속성은 Drop 하지 않음
df_titanic = df_titanic.drop(['Name', 'Ticket', 'Cabin'],axis=1)
df_titanic.head(5)
```

### 3-2) 결측치 확인 및 결측치 보간

```python
df_titanic.Age.fillna(df_titanic.Age.median(), inplace=True)
df_titanic.isnull().sum()
```

##### Embarked 의 Null값 제거

```python
df_titanic = df_titanic.dropna(subset='Embarked')
df_titanic.isnull().sum()
```

##### 결측치 없음

## 3-3) 원핫인코딩

```python
#인코딩 이전과 비교하기 위한 백업
dfbkup = df_titanic.copy()
```

```python
df_titanic = pd.get_dummies(df_titanic, columns = ['Embarked'])
df_titanic
```

```python
#[원핫인코딩 이후 데이터프레임]
X = np.array(df_titanic.iloc[:,2:]) 
Y = np.array(df_titanic['Survived'])
#[원핫인코딩 이전 데이터프레임]
X1 = np.array(dfbkup.iloc[:,2:]) 
Y1 = np.array(dfbkup['Survived']) 
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=0)
X1_train, X1_test, Y1_train, Y1_test = train_test_split(X1, Y1, test_size=0.3, random_state=0)

tree = DecisionTreeClassifier(random_state=12345)
tree1 = DecisionTreeClassifier(random_state=12345)

tree.fit(X_train, Y_train)
tree1.fit(X1_train, Y1_train)

print('[원핫인코딩 이후 모델 생성]')
print(classification_report(Y_test, tree.predict(X_test)))
print(confusion_matrix(Y_test, tree.predict(X_test)))

print('[원핫인코딩 없이 모델 생성]')
print(classification_report(Y1_test, tree1.predict(X1_test)))
print(confusion_matrix(Y1_test, tree1.predict(X1_test)))
```

##### 정확도에서는 원핫인코딩 전후로 차이가 없었으나, macro 평균에서는 오히려 약간 정확도가 하락한 것을 보았다. 탑승 항구에 따른 생존률의 차이는 없기 떄문이다.
